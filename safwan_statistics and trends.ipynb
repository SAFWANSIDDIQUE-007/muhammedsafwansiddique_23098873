```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.patches as mpatches

from collections import Counter
from typing_extensions import dataclass_transform
```

```python
df= pd.read_csv('/content/netflix_titles.csv')
df
```

```python
df.head()
```

```python
df.tail()
```

```python
df.drop([])
```

```python
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
df['year_added']=df['date_added'].dt.year
```

```python
df.describe(include='all')
```

```python
df['duration'] = df['duration'].str.extract('(\d+)').astype(float)
df.head()
```

```python
df.info()
```

```python
print(df.isnull().sum())
df=df.dropna(subset=['director'])
df=df.dropna(subset=['cast'])
df=df.dropna(subset=['country'])
df=df.dropna(subset=['date_added'])
df=df.dropna(subset=['rating'])
df=df.dropna(subset=['duration'])
df.info()

```

```python
print(df.isnull().sum())
df.info()


```

```python
duplicates=df.duplicated().sum()
print(f'Duplicates: {duplicates}')
```

```python
#Histogram plot
column = 'year_added'
plt.figure(figsize=(7, 5))
plt.hist(df[column], bins=13)
plt.title(f'NETFLIX CONTENT ADDITION BY YEARS')
plt.xlabel('Content Added Years')
plt.ylabel('Number of Titles Added')
mean = df[column].describe()['mean']
std = df[column].describe()['std']
plt.axvline(mean, color='red', linestyle='--', label=f'Mean: {mean:.2f}')
plt.axvline(mean + std, color='green', linestyle='--', label=f'+1 Std: {mean + std:.2f}')
plt.axvline(mean - std, color='yellow', linestyle='--', label=f'-1 Std: {mean - std:.2f}')
plt.xticks(np.linspace(2008, 2021, 14, dtype=int))
plt.legend()
plt.show()
```

```python
#Scatter plot
x_column = 'release_year'
y_column = 'duration'
movie_correlation = df[df['type'] == 'Movie'][[x_column, y_column]].corr().iloc[0, 1]
tv_show_correlation = df[df['type'] == 'TV Show'][[x_column, y_column]].corr().iloc[0, 1]
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x=x_column, y=y_column, hue='type')
plt.text(df[df['type'] == 'Movie'][x_column].mean()-55, df[df['type'] == 'Movie'][y_column].mean()+100,
         f'Movie Correlation: {movie_correlation:.2f}', fontsize=12, color='blue', ha='center')
plt.text(df[df['type'] == 'TV Show'][x_column].mean()-60, df[df['type'] == 'TV Show'][y_column].mean(),
         f'TV Show Correlation: {tv_show_correlation:.2f}', fontsize=12, color='orange', ha='center')
movie_patch = mpatches.Patch(color='blue', label=f'Movie Correlation: {movie_correlation:.2f}')
tv_show_patch = mpatches.Patch(color='orange', label=f'TV Show Correlation: {tv_show_correlation:.2f}')
legens1=plt.legend(handles=[movie_patch, tv_show_patch], loc='upper right')
plt.legend(loc='upper right', title='Type')
plt.title(f'SCATTER PLOT: RELEASE YEAR vs DURATION BY TYPE')
plt.xlabel(x_column)
plt.ylabel(y_column)
plt.show()
```

```python
#Heatmap plot
countries = df['country'].str.split(', ').apply(pd.Series).stack().reset_index(level=1, drop=True)
countries = countries.str.strip()
country_df = pd.get_dummies(countries)
country_df = country_df.groupby(level=0).sum()
top_10_countries = country_df.sum().sort_values(ascending=False).head(10).index
filtered_country_df = country_df[top_10_countries]
df_with_countries = pd.concat([df[['year_added']], filtered_country_df], axis=1)
df_with_countries['title_count'] = 1
heatmap_data = df_with_countries.groupby(['year_added'])[top_10_countries].sum().reset_index()
correlation_matrix = heatmap_data[top_10_countries].corr()
plt.figure(figsize=(7, 5))
sns.heatmap(correlation_matrix, cmap="coolwarm", annot=True, fmt=".2f")
plt.title("CORRELATION BETWEEN TOP 10 COUNTRIES AND NUMBER OF TITLES ADDED")
plt.xlabel("Country")
plt.ylabel("Country")
plt.tight_layout()
plt.show()
```
